```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("Data/mnist.csv")
x = df.drop(['label'], axis=1).values
y = df['label'].values
x_train, x_val, y_train, y_val = train_test_split(x,y,test_size=.25,random_state=43)
```

```python
from NNetwork import Fully_Connected
net = Fully_Connected() # create the neular net
net.init_layers_xavier(layers=[28*28,100,50,10]) # init the layers and size of layer 
net.train(x_train, y_train, x_val=x_val, y_val = y_val, num_iters=600, verbose=True, learning_rate=0.001) # start training 
```

```python
# you can adjust the learning rate and continue training 
net.train(x_train, y_train, x_val=x_val, y_val = y_val, num_iters=400, verbose=True, learning_rate=0.0001)
```

```python
plt.plot(net.loss_history)
plt.xlabel("Step")
plt.ylabel("Loss")
plt.show()
```

```python
plt.plot(net.train_acc_history  , 'b', label='traing accuracy')
plt.plot(net.val_acc_history  , 'r', label='validation accuracy')
plt.legend()
plt.show()
```
